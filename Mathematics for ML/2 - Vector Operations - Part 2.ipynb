```python
import numpy as np
```

Dot Product of 2 Vectors

```python
a = np.array([2,3])
b = np.array([4,4])

a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

```python
c = np.array([40,20,35])
d = np.array([53,24,68])
c_dot_d = np.dot(c,d)
```

```python
print(c_dot_d)
```

Cross Product of 2 Vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

Projection of "a" vector on "v" vector

```python
a = np.array([2,5])
v = np.array([8,-6])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print("Projection of a vector on v vector: ",proj_of_a_on_v)

```
